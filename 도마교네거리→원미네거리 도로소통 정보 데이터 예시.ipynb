```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline 
```

```python
csv = pd.read_csv('D:/download/프로젝트_download/지도/도로소통 정보 데이터/8_31_9_4_데이터.csv', encoding = "CP949")
csv.head()
```

```python
csv1 = pd.read_csv('D:/download/프로젝트_download/지도/도로소통 정보 데이터/9_7_검증데이터.csv', encoding = "CP949")
csv1.iloc[1:,:]
```

```python
plt.rc('font', family='Malgun Gothic')  # 한글깨짐 -> 맑은 고딕으로 설정
```

```python
plt.plot(data)
label = ['8월 31일 (월)','9월 1일  (화)','9월 2일  (수)','9월 3일  (목)','9월 4일  (금)','9월 7일  (월)']
plt.legend(labels = label)
plt.xticks(range(0,len(data.index)),data.index)
plt.grid()
plt.show()
```
